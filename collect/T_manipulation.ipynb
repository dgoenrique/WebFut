```python
import re

def extract_year(prod_nome):
  match = re.search(r'(\d{4}(?:-\d{2})?)', prod_nome)
  if match:
    return match.groups(1)
  return None
```

```python
nomes_produtos = [
    "Flamengo Home 2008-09 M/L",
    "Camisa Vasco 2000",
    "São Paulo Home 1995",
    "Corinthians Treino 2023",
    "Produto Sem Ano",
    "Grêmio Away 2010-11",
    "Cruzeiro 2005"
]

for nome in nomes_produtos:
  ano = extract_year(nome)
  print(f'Nome: {nome}, Ano: {ano}')
```

```python
def clean_product_name_year(prod_nome):
  pattern = r'\s*\d{4}(?:-\d{2})?.*$'

  cleaned_name = re.sub(pattern, '', prod_nome)
  return cleaned_name.strip()

```

```python
nomes_produtos = [
    "Real Madrid Home 2016-17 M/L",
    "Flamengo Home 2008-09",
    "Camisa Vasco 2000 P",
    "São Paulo Home 1995 G",
    "Corinthians Treino 2023 - Tamanho Único",
    "Produto Sem Ano Extra",
    "Grêmio Away 2010-11 GG",
    "Cruzeiro 2005 - Com Patrocínio",
    "Bayern de Munique Away 2020-21 XL Versão Jogador"
]

for i in nomes_produtos:
  print(clean_product_name_year(i))
```

```python
from ast import pattern
def product_type(prod_nome):
  pattern = r"(home|away|third)"
  match = re.search(pattern, prod_nome)
  if match:
    return match.group(1)
  return None
```

```python
nomes_produtos = [
    "Flamengo Home 2008-09",
    "Camisa Vasco 2000",
    "São Paulo Home 1995",
    "Corinthians Treino 2023",
    "Produto Sem Ano",
    "Grêmio Away 2010-11",
    "Cruzeiro 2005",
    "Real Madrid Third 2022/23",
    "Barcelona Kit #7",
    "Manchester United Away Jersey",
    "Camisa Seleção Edição Especial",
    "Camisa Flamengo Edição Limitada"
]

for i in nomes_produtos:
  print(product_type(i))
```

```python
def clean_product_name_type(prod_nome):
  pattern = r"(home|away|third)"
  cleaned_name = re.sub(pattern, '', prod_nome)
  return cleaned_name.strip()
```

```python
nomes_produtos = [
    "Flamengo Home 2008-09",
    "Camisa Vasco 2000",
    "São Paulo Home 1995",
    "Corinthians Treino 2023",
    "Produto Sem Ano",
    "Grêmio Away 2010-11",
    "Cruzeiro 2005",
    "Real Madrid Third 2022/23",
    "Barcelona Kit #7",
    "Manchester United Away Jersey",
    "Camisa Seleção Edição Especial",
    "Camisa Flamengo Edição Limitada" # Exemplo com "Edição"
]

for i in nomes_produtos:
  print(clean_product_name_type(i))
```

```python
def limpar_nome_time_camiseta(nome_completo):
    """
    Função para extrair o nome padronizado do time da string completa da camiseta.
    Precisa ser ajustada conforme a inconsistência dos seus dados.
    """
    if pd.isna(nome_completo): # Lida com valores NaN/nulos
        return None

    nome_completo_lower = str(nome_completo).lower()


    nome_limpo_temp = re.sub(r'\s*\d{4}(?:-\d{2})?.*$', '', nome_completo_lower).strip()


    nome_final_limpo = re.sub(r'\s*(?:home|away|third)\s*', '', nome_limpo_temp).strip()

    # --- Regras para nomes compostos ou específicos ---
    # Adicione suas regras específicas AQUI para garantir correspondência exata
    # com os nomes em df_times['Nome do Time']
    if 'real madrid' in nome_final_limpo:
        return 'real madrid'
    elif 'psg jordan' in nome_final_limpo or 'psg' in nome_final_limpo:
        return 'psg'
    elif 'atletico de madrid' in nome_final_limpo or 'atlético de madrid' in nome_final_limpo:
        return 'atlético de madrid'
    elif 'manchester united' in nome_final_limpo:
        return 'manchester united'
    elif 'bayern munich' in nome_final_limpo:
        return 'bayern munich'
    elif 'inter de milão' in nome_final_limpo:
        return 'inter de milão'
    elif 'boca juniors' in nome_final_limpo:
        return 'boca juniors'
    elif 'real betis' in nome_final_limpo:
        return 'real betis'
    elif 'olympique de marsellie' in nome_final_limpo:
        return 'olympique de marsellie'
    elif 'manchester city' in nome_final_limpo:
        return 'manchester city'
    elif 'borussia dortmund' in nome_final_limpo:
        return 'borussia dortmund'
    elif 'west ham' in nome_final_limpo:
        return 'west ham'
    elif 'river plate' in nome_final_limpo:
        return 'river plate'
    elif 'new york city' in nome_final_limpo:
        return 'new york city'
    elif 'celta de vigo' in nome_final_limpo:
        return 'celta de vigo'
    elif 'kaizer chiefs' in nome_final_limpo:
        return 'kaizer chiefs'
    elif 'irlanda do sul' in nome_final_limpo:
        return 'irlanda do sul'
    elif 'los angeles galaxy' in nome_final_limpo:
        return 'los angeles galaxy'
    elif 'corinthians' in nome_final_limpo:
        return 'corinthians'
    elif 'flamengo' in nome_final_limpo:
        return 'flamengo'
    elif 'brasil' in nome_final_limpo:
        return 'brasil'
    elif 'alemanha' in nome_final_limpo:
        return 'alemanha'


    # Adicione mais 'elif' para outros nomes compostos ou padrões que precisam de ajuste

    # Remover múltiplos espaços que podem ter sido criados
    return re.sub(r'\s+', ' ', nome_final_limpo).strip()
```

```python
camisetas['nome_time_padronizado'] = camisetas['nome'].apply(limpar_nome_time_camiseta)
```

```python
data_id = pd.merge(
    camisetas,
    data2,
    left_on='nome_time_padronizado',
    right_on='time',
    how='left'
)
data_id.head()
```

```python
df_camisetas_com_id_time = data_id.drop(
    columns=['time', 'time_id', 'nome_time_padronizado']
)
df_camisetas_com_id_time['id_time'] = df_camisetas_com_id_time['id']
df_camisetas_com_id_time = df_camisetas_com_id_time.drop(columns=['id'])
df_camisetas_com_id_time.head()
```
